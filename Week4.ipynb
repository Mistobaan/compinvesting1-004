# How to build a portfolio optimizer
## Inputs
- Expected Return for each equity
- volatility (risk) for each equity
- Target return 
- Covariance matrix (how much that equity affects other equities)


## Outputs
- Portfolio weights that minimize risk for target returns

Find a set of equities that have less risk than a single equity.

- correlation coefficient | from -1. to 1.0. 0 is random fashion
- covariance

# Efficient Frontier

You set the return target and you get the lower risk. Iterate over the return target.
The max return will be given from the single best performing equity in your portfolio.
You can set how max of each equity you want to have to limit this behavior.

usually we can optimize for the sharpe ratio to be in the middle of high returns and low risk

# Lecture 3: How an Optimzier Works ?
